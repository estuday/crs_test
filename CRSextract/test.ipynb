```python
from dao import add_knowledge
from dto import Knowledge
for i in range(len(data)):
    k_info =  data.iloc[i]
    _name_emb = requests.post(url,json={
    "texts": [
        k_info["名称"]
    ]
    })
    name_emb = json.loads(_name_emb.text)['data'][0]
    knowledge_dict = {"name":k_info["名称"],"description":k_info["介绍说明"],"level":k_info["知识点教学层次"],"type":k_info["知识点教学类型"],"difficulty":k_info["复杂度"],"name_emb":name_emb}
    if not isinstance(k_info["关键词"],float):
        knowledge_dict["keywords"] = k_info["关键词"]
    if not isinstance(k_info["别名"],float):
        knowledge_dict["alias"] = k_info["别名"]
    knowledge = Knowledge(
        **knowledge_dict
    )

    add_knowledge(knowledge=knowledge)
```
